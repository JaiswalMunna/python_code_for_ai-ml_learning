```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Data={'Tasks':[100,500,300]}
df=pd.DataFrame(Data,columns=['Tasks'],index=['Pending','Complete','Ongoing'])

df.plot.pie(y='Tasks',figsize=(5,5))
plt.show()
```

```python
labels_ex='JS','JAVA','Python','R'
sizes=[15,30,45,10]
explode_labels=(0,0.2,0,0)
fig1,ax1=plt.subplots()


ax1.pie(sizes,explode=explode_labels,labels=labels_ex,shadow=True,startangle=90)
ax1.axis('equal')
plt.show()
```
