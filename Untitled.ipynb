```python
def factorial(n):
    fact=1
    for i in range(1,n+1):
        fact=fact*i
    return fact
```

```python
factorial(5)
```

```python
user_input=int(input("Enter your number"))
factorial(user_input)
```

Sum of all the number till the number provided

```python
def natural(n):
    num=0
    for i in range(1,n+1):
        num=num+i
    return num
```

```python
natural(5)
```

```python
user_input=int(input("Enter your number"))
natural(user_input)
```
