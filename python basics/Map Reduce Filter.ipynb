####Area of Circle

```python
import math

def area(r):
    return math.pi*(r**2)

radii=[1,2,4,5,7,8]

areas=[]

for i in radii:
    a=area(i)
    areas.append(a)

print(areas)
```

Using map function:-  map(f,iterable)

```python
map(area, radii) #it is complete and to call this there is a method below
list(map(area, radii))
```

Convert Celcius scale to Farheneit scale

F=9/5*C+32

```python
temps=[("Mumbai",32),("Delhi",42),("Chennai",25),("Bihar",38)]

cel_to_fra=lambda data: (data[0],(9/5)*data[1]+32)

map(cel_to_fra,temps)

list(map(cel_to_fra,temps))
```

Filter Function- filters out the data

filter(f, data)

```python
import statistics
```

```python
data=[54,65,8,4,2,1,4,6,5,4,68,7,4,6,8]
```

```python
avg=statistics.mean(data)
```

```python
print(avg)
```

to print more than avg number in above data

```python
list(filter(lambda x: x>avg, data))
```
