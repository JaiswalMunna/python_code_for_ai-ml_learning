```python
#nested list
city_name=["india",["Delhi","Mumbai"],"Nepal",["kathmandu","pokhara"]]
```

```python
print(city_name)
print(type(city_name))
```

```python
print(city_name[0])
print(city_name[1])
print(city_name[2])
print(city_name[3])
```

```python
print(city_name[1][0])
print(city_name[1][1])
print(city_name[3][0])
print(city_name[3][1])
```

```python
#access from backward direction
print(city_name[-1][-1])
print(city_name[-1][1])
print(city_name[-1][-2])
```

```python
list1=[1,2,3,4,5]
print(list1)
del list1[3]
print(list1)
```

```python
print(list1)
del list1
print(list1)  #using del keyword we can delete wholelist or item of the list
```

```python
#or we can use pop it work inindex based. if we don't pass index then it will remove or delete the last item of the list
list1=[1,2,3,4,5]
list1.pop(3)
print(list1)
list1.pop()
print(list1)
```

```python
#remove willn't work as a index element of the list. it chek the number that we pass in remove, 
#if that number present then it will remove that number or give an error. 
#and if we use remove without index then it will delete the whole list
list1=[1,92,3,86,64,5]
list1.remove(4)
print(list1)
```
