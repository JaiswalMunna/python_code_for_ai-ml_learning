Defining a Class

Default Constructor

```python
class Rectangle:
    def __init__(self):
        self.length=10
        self.breadth=5
```

```python
rect=Rectangle()
rect.breadth
```

```python
rect=Rectangle()
print("Length: ",rect.length,"\nBreadth:",rect.breadth)
```

Parametrised Constructor

```python
class Rectangle:
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
```

```python
rect=Rectangle(20,15)
```

```python
print("Length: ",rect.length,"\nBreadth:",rect.breadth)
```

Class variables and Instance Variables

```python
class Circle: 
    pi=3.14159
    def __init__ (self,radius):
        self.radius= radius
```

```python
#instantiating the circle class
c=Circle(5)
print("radius=",c.radius)
print("pi=",c.pi)

Circle.pi=3.14444
print("pi=",c.pi)
```

Calculate a area of rectangle

```python
class Rectangle:
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
        
    def calculate_area(self):
        return self.length*self.breadth
```

```python
rect=Rectangle(10,20)
print("Area of rectangle is:",rect.calculate_area())
```

Inheritance and Overriding

```python
class Employee:
    def function(self):
        print("Hello World")

class Department(Employee):
    pass
```

```python
emp=Employee()
emp.function()
```

```python
dept=Department()
dept.function()
```

```python
class Shape:
    def set_color(self,color):
        self.color=color
        
    def calculate_area(self):
        pass
        
    def color_the_shape(self):
        color_price={"red":10,"blue":15,"green":5}
        return self.calculate_area()*color_price[self.color]
```

```python
class Circle(Shape):
    pi=3.14159
    def __init__(self,radius):
        self.radius=radius
    
    def calculate_area(self):
        return Circle.pi*self.radius
    
```

```python
c=Circle(10)
#Accessing set_color from parent class, because of iniheritance
c.set_color("red")
print("Circle with radius =", c.radius,"When colored",c.color,"Costs $",c.color_the_shape())
```
