```python
x=float(input("enter the number"))
print(x)
if x%2==0:
    print(x, "is an even number")
else:
    print(x, "is an odd number")
```

```python
x=int(input("enter your age"))
print(x)
if x>=18 & x>50:
    print("Your age is ",x, "\nSo, you can join the party.")
else:
    print("Your age is ",x, "\nSo, you can't join the party.")
```

Iterations

```python
list1=[10,20,30,40,50]
print(list1)
len(list1)
```

```python
#iterate over a list of integers
list1=[10,20,30,40,50]
for i in list1:
    print(i)
    i=i+1
```

```python
#iterate over a dict
student_data={1:["Munna",25],2:["Priyanka",22],3:["Munnu",20],4:["Naushad",27],5:["SHyam",25]}
for i in student_data.items():
    print(i)
```

```python
#iterate over a dict elements
student_data={1:["Munna",25],2:["Priyanka",22],3:["Munnu",20],4:["Naushad",27],5:["SHyam",25]}
for i,j in student_data.items():
    print(j)
```

```python
#iterate over a dict keys
student_data={1:["Munna",25],2:["Priyanka",22],3:["Munnu",20],4:["Naushad",27],5:["SHyam",25]}
for i,j in student_data.items():
    print(i)
```
